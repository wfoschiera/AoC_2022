## Reading data

```python
with open("input.txt", 'r') as file:
    data = file.read()
print(data)
```

## Cleaning

```python
calories_per_elf_raw = data.split('\n\n')
calories_per_elf = []
for elf in calories_per_elf_raw:
    calories_str = elf.split('\n')
    calories_per_elf.append([int(cal) for cal in calories_str])
print(calories_per_elf)
```

## Finding the answer
Find the Elf carrying the most Calories. How many total Calories is that Elf carrying?

```python
max_calories = 0
total_calories = []

for elf in calories_per_elf:
    total = sum(elf)
    if total > max_calories:
        max_calories = total
    total_calories.append(total)
print(max_calories)
    
```

## Part 2
Find the top three Elves carrying the most Calories. How many Calories are those Elves carrying in total?

```python
result = sorted(total_calories)
sum(result[-3:])
```
