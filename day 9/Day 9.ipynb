```python
with open("input.txt", 'r') as txt_input:
  data = txt_input.read().splitlines()
print(data)
```

```python
def is_attached(head_position, tail_position):
    return abs(head_position[0] - tail_position[0]) < 2 and abs(head_position[1] - tail_position[1]) < 2
```

```python
def move_tail(head_position, tail_position):
    dif_x = head_position[0] - tail_position[0]
    dif_y = head_position[1] - tail_position[1]
    if dif_x > 0:
        if dif_y > 0:
            return (tail_position[0]+1, tail_position[1]+1)
        if dif_y < 0:
            return (tail_position[0]+1, tail_position[1]-1)
        if dif_y == 0:
            return (tail_position[0]+1, tail_position[1])
    if dif_x < 0:
        if dif_y > 0:
            return (tail_position[0]-1, tail_position[1]+1)
        if dif_y < 0:
            return (tail_position[0]-1, tail_position[1]-1)
        if dif_y == 0:
            return (tail_position[0]-1, tail_position[1])
    if dif_x == 0:
        if dif_y > 0:
            return (tail_position[0], tail_position[1]+1)
        if dif_y < 0:
            return (tail_position[0], tail_position[1]-1)
```

### Testing

```python
assert move_tail((1,2), (0,0)) == (1,1)
assert move_tail((0,0), (1,2)) == (0,1)
assert move_tail((1,0), (0,2)) == (1,1)
assert move_tail((0,2), (1,0)) == (0,1)
```

```python
def move_head(head_position, direction):
    match direction:
        case "R":
            head_position = (head_position[0] + 1, head_position[1])
        case "L":
            head_position = (head_position[0] - 1, head_position[1])
        case "U":
            head_position = (head_position[0], head_position[1] + 1)
        case "D":
            head_position = (head_position[0], head_position[1] - 1)
    return head_position
```

### Testing

```python
assert move_head((0,0), "R") == (1,0)
assert move_head((1,1), "L") == (0,1)
assert move_head((0,0), "U") == (0,1)
assert move_head((1,1), "D") == (1,0)
```

```python

tail_tracking = [(0,0)]
def wrapper(data):
    """makes 1 move at time. Move head and than move tail. 
    return head_position, tail_position"""
    head_position = (0,0)
    tail_position = (0,0)
    
    for line in data:
        direction, size = line.strip().split(" ")
        for i in range(int(size)):
            head_position = move_head(head_position, direction)
            if is_attached(head_position, tail_position):
                continue 
            tail_position = move_tail(head_position, tail_position)
            tail_tracking.append(tail_position)

    return tail_tracking
```

```python
tail_tracking = wrapper(data)
```

```python
def wrapper_part_2(data):
    tail_tracking = []
    knots_pos = [(0,0) for _ in range(10)]
    next_knot = (0,0)
    
    for line in data:
        direction, size = line.strip().split(" ")
        
        for i in range(int(size)):
            knots_pos[0] = move_head(knots_pos[0], direction)

            for i, knot in enumerate(knots_pos):
                if i == 0:
                    continue
                    
                if not is_attached(knots_pos[i-1], knot):
                    knots_pos[i] = move_tail(knots_pos[i-1], knot)
                    tail_tracking.append(knots_pos[-1])

        print("Knots position: ", knots_pos)
        print("------")
    return tail_tracking
```

```python
tail_tracking = wrapper_part_2(data)
```
